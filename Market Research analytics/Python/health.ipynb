# handling missing values

```python
health['outbreak_starting_date'].isnull().sum()
```

```python
health['outbreak_starting_date']
```

```python
health['outbreak_starting_date'] = pd.to_datetime(
    health['outbreak_starting_date'],
    dayfirst=True,
    errors='coerce'
)
```

```python
health[health['outbreak_starting_date'].isna()]
```

```python
health['outbreak_starting_date'].isna().sum()
```

```python
health['state'].isnull().sum()
```

```python
health['state'].value_counts()
```

```python
health['state']=health['state'].str.lower()
health['state']=health['state'].str.title()
```

```python
health['state'] = health['state'].replace({'Daman And Diu':'Dadra And Nagar Haveli And Daman And Diu',
'Dadra And Nagar Haveli':'Dadra And Nagar Haveli And Daman And Diu'} )
```

```python
health['state'] = health['state'].replace({'Arunachal':'Arunachal Pradesh','Madhya':'Madhya Pradesh','Uttar':'Uttar Pradesh'})
```

```python
health['state'].unique()
```

```python
health['state']=health['state'].fillna('Unknown')
```

```python
health['district'].describe()
```

```python
health['district']=health['district'].str.lower()
health['district']=health['district'].str.title()
health['district']=health['district'].str.strip()
```

```python
health['district'] =( health['district'].str.split().str.join(" "))
```

```python
health['district'].isna().sum()
health['district'].value_counts().head()
```

```python
health['district']=health['district'].fillna("Unknown")
```

```python
health['district']=health['district'].replace({'badaun':'Budaun','Baudaun':'Budaun','Thiruvallur':'Tiruvallur'})
```

```python
health['district'].unique()
```

```python
health['disease_illness_name']=health['disease_illness_name'].fillna("Unknown")
```

```python
health['disease_illness_name'].describe()
```

```python
health['disease_illness_name'].value_counts()
```

```python
health['disease_illness_name']=health['disease_illness_name'].str.lower()
health['disease_illness_name']=health['disease_illness_name'].str.title()
```

```python
health['disease_illness_name'].unique()
```

```python
health['status'].unique()
```

```python
health['status']=health['status'].str.lower()
health['status']=health['status'].str.title()
```

```python
health['status'].unique()
```

```python
health['cases'].describe()
```

```python
health['cases']=health.groupby("district")['cases'].transform(lambda x:x.fillna(x.median()))
```

```python
health['cases']=health['cases'].astype(int)
```

```python
health['deaths']=health.groupby('district')['deaths'].transform('mean')
```

```python
health['deaths']=health['deaths'].astype(int)
```
