# DATA PREPROCESSING

```python
aqi.columns
```

```python
aqi.columns=aqi.columns.str.lower()
aqi.columns=aqi.columns.str.replace(" ","_")
```

```python
aqi.info()
```

```python
aqi['aqi_value']=aqi['aqi_value'].astype(int)
```

```python
aqi.isnull().sum()
```

```python
aqi['number_of_monitoring_stations']=aqi['number_of_monitoring_stations'].fillna(aqi['number_of_monitoring_stations'].mean())
```

```python
aqi['number_of_monitoring_stations']=aqi['number_of_monitoring_stations'].astype(int)
```

```python
aqi['prominent_pollutants'].head()
```

```python
aqi=aqi.dropna(subset=['prominent_pollutants'])
```

```python
aqi['prominent_pollutants'].head()
```

```python
aqi.isnull().sum()
```

```python
aqi = aqi.drop(columns=['unit'])
```

```python
aqi = aqi.drop(columns=['note'])
```

```python
aqi['date'] = pd.to_datetime(aqi['date'], format='%d-%m-%Y')
```

```python
aqi = aqi.rename(columns={'area': 'city'})
```
